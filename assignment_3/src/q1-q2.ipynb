## Noise removal

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img1 = cv2.imread('./img/fce(salt_pepper noise).bmp')
img2 = cv2.imread('./img/Copy of Lena-Gaussian-noise2.jpg')

gray_img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
gray_img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(gray_img1, 'gray')
plt.title("salt-pepper noise")
plt.axis("off")

plt.subplot(122)
plt.imshow(gray_img2, 'gray')
plt.title("gaussian noise")
plt.axis("off")
plt.show()
```

### Average filtering

```python
plt.figure(figsize=(12,12))
for i, ksize in enumerate([1, 2, 3, 4, 5, 6, 7, 8, 9]):
    avg = cv2.blur(gray_img2, (ksize,ksize))
    plt.subplot(3,3,i+1), plt.title(f"average: {ksize}x{ksize}"), plt.imshow(avg, cmap='gray'), plt.axis("off")
plt.show()

plt.figure(figsize=(12,16))
for i, sig in enumerate([0, 1]):
    for j, ksize in enumerate([1, 3, 5, 7, 9, 11]):
        gau = cv2.GaussianBlur(gray_img2, (ksize,ksize), sig)
        plt.subplot(4,3,i*6+j+1), plt.title(f"gaussian: {ksize}x{ksize}, {sig}"), plt.imshow(gau, cmap='gray'), plt.axis("off")
plt.show()

plt.figure(figsize=(12,12))
for i, ksize in enumerate([1, 3, 5, 7, 9, 11]):
    med = cv2.medianBlur(gray_img2, ksize)
    plt.subplot(3,3,i+1), plt.title(f"median: {ksize}x{ksize}"), plt.imshow(med, cmap='gray'), plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.subplot(131)
plt.imshow(cv2.blur(gray_img2, (3,3)), 'gray')
plt.title("avgerage: 3x3")
plt.axis("off")

plt.subplot(132)
plt.imshow(cv2.GaussianBlur(gray_img2, (5,5), 1), 'gray')
plt.title("gaussian: 5x5, 1")
plt.axis("off")

plt.subplot(133)
plt.imshow(cv2.medianBlur(gray_img2, 3), 'gray')
plt.title("median: 3x3")
plt.axis("off")
plt.show()
```
