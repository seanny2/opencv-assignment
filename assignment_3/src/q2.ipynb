```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('./img/flower.bmp')

gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(gray_img, 'gray')
plt.axis("off")
```

### 에지 검출 - sobel

```python
# Sobel operator
x = cv2.Sobel(gray_img, -1, 1, 0, delta=128)
y = cv2.Sobel(gray_img, -1, 0, 1, delta=128)

# Turn uint8, image fusion
absX = cv2.convertScaleAbs(x)
absY = cv2.convertScaleAbs(y)
output_img = cv2.addWeighted(absX, 0.7, absY, 0.1, 0)

plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(gray_img, 'gray')
plt.title("original")
plt.axis("off")
plt.subplot(122)
plt.imshow(output_img, 'gray')
plt.title("sobel")
plt.axis("off")
plt.show()
```

### 에지 검출 - prewitt

```python
# 프르윗 필터
gx = np.array([[-1,0,1],[-1,0,1],[-1,0,1]],dtype=int)
gy = np.array([[1,1,1],[0,0,0],[-1,-1,-1]],dtype=int)

# 프르윗 필터 컨벌루션
x = cv2.filter2D(gray_img, -1, gx, delta=128)
y = cv2.filter2D(gray_img, -1, gy, delta=128)

# uint8 타입(0~255)로 변경하고 영상 합하기
absX = cv2.convertScaleAbs(x)
absY = cv2.convertScaleAbs(y)
output_img = cv2.addWeighted(absX, 0.7, absY, 0.1, 0)

plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(gray_img, 'gray')
plt.title("original")
plt.axis("off")
plt.subplot(122)
plt.imshow(output_img, 'gray')
plt.title("prewitt")
plt.axis("off")
plt.show()
```

### 에지 검출 - canny

```python
# 케니 에지 컨벌루션 연산하기
output_img = cv2.Canny(gray_img,240,250)

plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(gray_img, 'gray')
plt.title("original")
plt.axis("off")
plt.subplot(122)
plt.imshow(output_img, 'gray')
plt.title("canny")
plt.axis("off")
plt.show()
```

### 에지 검출 - robert cross

```python
# 로버트 크로스 필터
gx = np.array([[-1, 0], [0, 1]], dtype=int)
gy = np.array([[0, -1], [1, 0]], dtype=int)

# 로버트 크로스 컨벌루션
x = cv2.filter2D(gray_img, -1, gx, delta=128)
y = cv2.filter2D(gray_img, -1, gy, delta=128)

# 절대값 취하기
absX = cv2.convertScaleAbs(x)
absY = cv2.convertScaleAbs(y)
output_img = cv2.addWeighted(absX, 0.7, absY, 0.1, 0)

plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(gray_img, 'gray')
plt.title("original")
plt.axis("off")
plt.subplot(122)
plt.imshow(output_img, 'gray')
plt.title("robert cross")
plt.axis("off")
plt.show()
```
