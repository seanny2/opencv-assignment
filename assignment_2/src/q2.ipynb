```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

origin_img = cv2.imread('./img/peppers.jpg')

# BGR채널순서를 RGB채널로 변경
RGB_img = cv2.cvtColor(origin_img, cv2.COLOR_BGR2RGB) 

# RGB 채널 나누기
Red_img, Green_img, Blue_img = cv2.split(RGB_img)

plt.figure(figsize=(14,10))

plt.subplot(221)
plt.imshow(RGB_img)
plt.title("original image")
plt.axis("off")

# 히스토그램 출력, 채널은 0으로 표시
plt.subplot(222)
hist = cv2.calcHist([Red_img], [0], None, [256], [0, 256])
plt.xlim([0, 256])
plt.plot(hist, color = 'r') # R 색상 히스토그램 표현

hist = cv2.calcHist([Green_img], [0], None, [256], [0, 256])
plt.xlim([0, 256])
plt.plot(hist, color = 'g') # G 색상 히스토그램 표현

hist = cv2.calcHist([Blue_img], [0], None, [256], [0, 256])
plt.xlim([0, 256])
plt.plot(hist, color = 'b') # B 색상 히스토그램 표현
plt.show()

plt.figure(figsize=(15,10))

plt.subplot(131)
ret, Red_Binary_img = cv2.threshold(Red_img, 156, 255, cv2.THRESH_BINARY_INV)
plt.imshow(Red_Binary_img, 'gray')
plt.title("red image threshold = 90")
plt.axis("off")

plt.subplot(132)
ret, Green_Binary_img = cv2.threshold(Green_img, 100, 255, cv2.THRESH_BINARY)
plt.imshow(Green_Binary_img, 'gray')
plt.title("green image threshold = 100")
plt.axis("off")

plt.subplot(133)
ret, Blue_Binary_img = cv2.threshold(Blue_img, 52, 255, cv2.THRESH_BINARY)
plt.imshow(Blue_Binary_img, 'gray')
plt.title("blue image threshold = 52")
plt.axis("off")

plt.show()
```

```python
# 그레이스케일 이미지로 변환
Gray_img = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)

# 그레이스케일 이미지에 이진화
# ret: 설정 임계값, Binary_img: 결과 이미지
ret, Binary_img = cv2.threshold(Gray_img, 127, 255, cv2.THRESH_BINARY)

# gray-scale
plt.figure(figsize=(20, 5))
plt.subplot(131)
plt.title("gray-scale image")
plt.axis("off")
plt.imshow(Gray_img, 'gray')

plt.subplot(132)
hist = cv2.calcHist([Gray_img], [0], None, [256], [0, 256])
plt.xlim([0, 256])
plt.xticks([i*16 for i in range(17)])
plt.plot(hist)

plt.subplot(133)
plt.axis("off")
plt.title("threshold = 127")
plt.imshow(Binary_img, 'gray')

plt.tight_layout()
plt.show()
```
